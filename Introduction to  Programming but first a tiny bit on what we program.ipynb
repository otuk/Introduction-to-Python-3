# Programming and Computers

This is an introduction on to how to program computers.  Programming is also simply referred to as coding or even as hacking more and more.

First, we should learn a bit about the machine that we would like to program.  This is similar to knowing about the basics of how an airplane physically flies before learning to become a pilot.  It will help with the controls during the flight.



The hardware functioning of the computer, its basic internal architecture,  has not changed much at a very high level since its invention.  The important parts for programming are the CPU, RAM and the storage medium such as an HDD/SSD drive or a flash card.  The computer is still the same engine that is only capable of obeying what it is told to do, executing its instructions line by line to the dot, exactly as it is told.

How the computer is told what to do is the coding the rest of these notes will talk about.  The code is written in a file, then stored on a storage disk.  When you want to run your code, the code first gets loaded from the storage medium in to the RAM - Random Access Memory, and sits there ready for excution.  The piece that loads the code or data to the RAM is called the CPU - Central Processing Unit.  The CPU has a very close relationship with the RAM and is at the center of the computer, both a manager of what should be done next  but also the worker that does the work.  The CPU is the piece of hardware that is capable of moving data into and out of RAM.  It is also the piece that is capable of code execution.  Once the code is loaded into RAM, the CPU starts execution. The CPU will follow the programmer's instructions line by line and complete execution when the instructions end with the last line of code.



Of course in order to interact with the CPU/RAM and the storage of the computers, we humans need a keyboard to write on, a monitor to see, and a mouse to click. These are really not part of the core system, they are at the periphery of it and for that reason they are generally referred to as peripherals.  Everything mentioned so far is purchased under one name and in one convinient package; a laptop, a notebook or simply a computer.

There is one more thing to explain on the magic of the computer or how anything actually happens inside the computer: the concept of the clock inside the computer.  Given that all CPU, RAM and other components are very static, they are all hardwired and they just sit inside the computer without any moving parts, asking how does  anything actually happen or change is a valid question. That is due to  the one thing that changes non-stop at regular intervals inside a computer as soon as it is powered on until it is powered off. This is the clock pulse.  Think of it as never ending sequences of regularly spaced waves inside the wires of the computer.  The changes in the clock pulse tell the CPU that it is now time to execute the next instruction line in code and move the data generated by the previous instruction to whereever it needs to go.  It is this clock pulse that everything in the computer synchronizes themselves to, to make the code run as the programmer intended in a sequence of steps. Basically, it is the clock pulse that allows the execution of code line by line.  Not very different than passage of time allowing things to "happen" in the universe.

For a person starting to program, this level of how the hardware of the computer works might be sufficient for now.  More details on hardware  will be reviewed in later notes.
